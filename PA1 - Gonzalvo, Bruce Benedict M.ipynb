```python
def alphabet_soup(s: str) -> str:  #creates a function named 'alphabet_soup' with s as the string
    return ''.join(sorted(s))    #sort the characters of the strings in alphabetic order 
```

### Function Test

```python
print(alphabet_soup("hello"))
print(alphabet_soup("hacker"))
print(alphabet_soup("CLASS"))
```

```python
EMOTICONS = {"smile": ":)", "grin": ":D" , "sad": ":((" , "mad": ">:("}

def emotify(txt : str) -> str:   #text is a string
    return " ".join(EMOTICONS.get(word.lower(), word) for word in txt.split())
```

### Function Test

```python
print(emotify("make me Smile"))
print(emotify("Grin"))
print(emotify("Sad"))
print(emotify("Mad"))
```
